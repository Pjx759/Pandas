# 读取数据

```python
import pandas as pd
```

```python
df_raw = pd.read_csv("fsnd_zb_data.tsv",encoding="utf8",sep="\t", \
                        keep_default_na=False,na_values="na_rep",index_col=[0,1,2])
df_m = pd.read_csv ("fsnd_zb_meta1.tsv",encoding="utf8",sep="\t",keep_default_na=False,na_values="na_rep",index_col=[0])
df_r = pd.read_csv ("reg_treeId_level2.tsv",encoding="utf8",sep="\t",keep_default_na=False,na_values="na_rep")
display(df_r.tail())
# df_jy = pd.read_csv ("就业切片.tsv", encoding = "utf-8", sep ="\t")
```

## 创建相关字典

```python
# 创建指标字典
指标字典 = df_m['cname'].to_dict()
指标字典
```

```python
# 创建地区字典
地区字典 = df_r.set_index("id")['name'].to_dict()
地区字典
```

```python
df_raw.reset_index()
```

```python
# 指标字典建立->zb指标
df = df_raw.reset_index().set_index("zb").rename(index=指标字典)
df
```

```python
# 
# reg->地区
df = df.reset_index().set_index("reg").rename(index=地区字典)
df
```

```python
# 地区字典重命名
df_zh = df.reset_index().rename(columns={"reg":"地区","zb":"指标",\
                                         "sj":"年","data":"数据",})
df_zh
```

# 数据切片

```python
# 切数据
dslice = df_zh [ df_zh.指标.str.contains("城镇单位就业人员") ]
dslice
```

```python
# 查看
指标分的可能性 = dslice.指标.unique()
指标分的可能性
```

```python
# 分的可能性
指标分的可能性 = [ x.split("城镇单位就业") for x in dslice.指标.unique() ]
指标分的可能性
```

```python
# 取值
指标分的可能性_取 = [ [x,y] for (x,y) in 指标分的可能性 if (y=='人员平均工资' or y=='人员') and x !='' ]
指标分的可能性_取
```

```python
# 进一步切片
指标分的可能性_取_all = ["城镇单位就业".join(x) for x in 指标分的可能性_取]
指标分的可能性_取_all
```

```python
df_就业切片 = df_zh.set_index("指标").loc[指标分的可能性_取_all].reset_index()
df_就业切片
```

```python
df_就业切片.describe(include="all")
```

# 分组数据

```python
# 分进合击分组数据
df_就业切片.groupby(by='指标').agg({"数据":["min","mean","max"]})
```

```python
# 多层次数据
df_就业切片['行业'] = [x.split("城镇单位就业")[0] for x in df_就业切片.指标]
df_就业切片['行业指标'] = [x.split("城镇单位就业")[1] for x in df_就业切片.指标]
df_就业切片
```

```python
df_就业切片.groupby(['地区','行业','行业指标']).mean()
```

```python
df_就业切片.groupby(['地区','行业','行业指标']).agg({"数据":["min","mean","max"]})
```

```python
# 多层次数据报表输出
df_就业切片
df_就业切片.groupby(['行业','行业指标']).agg({"数据":["min","mean","max"]})
```

```python
df_就业切片.groupby(['行业','地区']).agg({"数据":["min","mean","max"]})
```
